```python
from sklearn.decomposition import TruncatedSVD
from scipy.sparse.linalg import svds
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


rating_data = pd.read_csv("completion/ratings_final.csv")
movie_data = pd.read_csv("completion/movie_final.csv")


movie_data.drop('genre', axis=1, inplace=True)
user_movie_data = pd.merge(rating_data, movie_data, on='content_id')

user_movie_rating = user_movie_data.pivot_table('rating', index = 'userId', columns='title').fillna(0)

movie_user_rating = user_movie_rating.values.T

SVD = TruncatedSVD(n_components=12)
matrix = SVD.fit_transform(movie_user_rating)

corr = np.corrcoef(matrix)
corr2 = corr[:200, :200]
plt.figure(figsize=(16, 10))
sns.heatmap(corr2, cmap=sns.color_palette('Greys', n_colors=10))
```
